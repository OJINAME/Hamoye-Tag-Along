## Modelling

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = data.drop(columns= ['target', 'stabf'] )
y = data[ 'target' ]
```

```python
s = StandardScaler()
```

```python
Xs = s.fit_transform(X)
```

```python
Xs = pd.DataFrame(Xs, columns=X.columns)
Xs.head()
```

```python
#split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Xs, y, test_size= 0.2 , random_state= 1 )
y_train.value_counts()
```

### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand_f = RandomForestClassifier(random_state=1)
rand_f.fit(x_train, y_train)
```

### Stratified K-Fold Cross Validation On Random Forest Classifier

```python
from sklearn.metrics import recall_score, accuracy_score, precision_score, f1_score, confusion_matrix
from sklearn.model_selection import StratifiedKFold
```

```python
folds = StratifiedKFold(n_splits=5)
```

```python
from sklearn.model_selection import KFold

fold = 0
F1_score = []
for train_index, val_index in folds.split(Xs,y):
    x_tr = Xs.loc[train_index]
    y_tr = y.loc[train_index]
    
    x_val = Xs.loc[val_index]
    y_val = y.loc[val_index]
    
    
    rf = RandomForestClassifier(random_state=1)
    rf.fit(x_tr, y_tr)
    pred = rf.predict(x_val)
    fscore = f1_score(y_val, pred)
    
    print(f"======= Fold {fold} ========")
    print(f"F1 Score on validation set is {fscore:0.4f}")
    
    fold += 1
    F1_score.append(fscore)
    
avg_F1_score = np.mean(F1_score)
print(f"Mean F1 Score score is {avg_F1_score:0.4f}")
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from matplotlib import pyplot as plt
from sklearn import metrics
import itertools
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
rand_pred = rand_f.predict(x_test)
print('Accuracy of RF Classifier on training set: {:.2f}'.format(rand_f.score(x_train, y_train)))
print('Accuracy of RF Classifier on test set: {:.2f}'.format(rand_f.score(x_test, y_test)))
```

```python
print(classification_report(y_test, rand_pred))
```

```python
cnf_mat = confusion_matrix(y_true=y_test, y_pred=rand_pred)
cnf_mat
```

```python
score = metrics.accuracy_score(y_test, rand_pred)
print(f'Accuracy: {round(score*100,2)}%')
cm = metrics.confusion_matrix(y_test, rand_pred)
plot_confusion_matrix(cm, classes=['Unstable', 'Stabe'])
```
